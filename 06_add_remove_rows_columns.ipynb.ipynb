# Adding and Removing Rows and Columns in Pandas

## Objective
This notebook demonstrates how to modify DataFrame structure by adding and removing rows and columns using vectorized operations and built-in pandas methods.

```python
people = {
    "first": ["Kenan", "Sercem" , "Ahmet", "Huseyin"],
    "last": ["KURT" , "MAKUM" , "AYCAN", "KURT"],
    "email":["kt@gmail.com", "sm@gmail.com" , "aa@gmail.com", "hd@gmail.com"]
}

import pandas as pd
df = pd.DataFrame(people)
df
```

```python
df['first'] + ' ' + df['last']
```

```python
df['full_name'] = df['first'] + ' ' + df['last']
df
```

```python
df.drop(columns=['first','last'], inplace = True)


```

```python
df
```

```python
df['full_name'].str.split(' ', expand = True)
```

```python
df[['first', 'last']] = df['full_name'].str.split(' ', expand = True)
df
```

```python
df.drop(index = 3)
```

```python
df
```

```python
df.drop(index = df[df['last'] == 'KURT'].index)
```

```python
people = {
    "first": ["Kenan", "Sercem" , "Ahmet", "Huseyin"],
    "last": ["KURT" , "MAKUM" , "AYCAN", "KURT"],
    "email":["kt@gmail.com", "sm@gmail.com" , "aa@gmail.com", "hd@gmail.com"]
}

import pandas as pd
df = pd.DataFrame(people)
df
```

```python
df = df[df['last'] != 'KURT']
df
```
