# Airbnb Data Cleaning with Pandas

This notebook focuses on cleaning an Airbnb listings dataset using Python and Pandas.

The cleaning process includes:

- Standardizing column names
- Fixing incorrect data types
- Cleaning currency fields (price, service_fee)
- Handling missing values
- Removing duplicate listings
- Filtering unrealistic values (minimum_nights, availability_365)
- Converting date columns

The final dataset contains **72,895 rows and 24 columns** and is ready for exploratory data analysis.

## 1. Import Libraries
## 2. Load Dataset

```python
import pandas as pd
data = pd.read_csv("Airbnb_Open_Data.csv", low_memory=False)
```

```python
## 2. Load Dataset
data
```

```python
data.columns
```

## 3. Column Name Cleaning

```python
data.columns = data.columns.str.strip().str.replace(' ','_')
data.columns = data.columns.str.lower()
data.columns
```

```python
data.dtypes
```

## 4. Data Type Corrections

```python
data['instant_bookable'] = data['instant_bookable'].astype('bool')
data['construction_year'] = data['construction_year'].astype('Int64')
data['price'] = data['price'].str.strip().str.replace('$','',regex=False).str.replace(',','').astype('Int64')
data['service_fee'] =  data['service_fee'].str.strip().str.replace(',','').str.replace('$','',regex=False).astype('Int64')
data['minimum_nights'] = data['minimum_nights'].astype('Int64')
data['calculated_host_listings_count'] = data['calculated_host_listings_count'].astype('Int64')
data['availability_365'] = data['availability_365'].astype('Int64')
data['last_review'] = pd.to_datetime(data['last_review'], format = '%m/%d/%Y')
data['number_of_reviews'] = data['number_of_reviews'].astype('Int64')
```

```python
data.info()
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns = ['license'], inplace = True)
data.drop(columns = ['house_rules'], inplace = True)
```

```python
data.isnull().sum()
```

## 5. Handling Missing Values

```python
data = data.dropna(subset=[
'price',
'service_fee',
'minimum_nights',
'availability_365',
'lat',
'long',
'neighbourhood',
'neighbourhood_group'
])
```

```python
data.isnull().sum()
```

## 6. Removing Duplicates

```python
data['id'].duplicated().sum()
```

```python
data = data.drop_duplicates(subset = ['id'])
```

## 7. Outlier Detection and Cleaning

```python
data['minimum_nights'].describe()
```

```python
data = data[(data['minimum_nights']>=1) & (data['minimum_nights']<=30)] 
```
